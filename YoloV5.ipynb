```python
import torch

# Load YOLOv5s model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
```

```python
import os
from PIL import Image, UnidentifiedImageError
```

```python
# Directory containing images
img_dir = '/content/datasets/coco128/images/train2017'
```

```python

# Get all image files in the directory
image_files = [os.path.join(img_dir, img) for img in os.listdir(img_dir) if img.endswith(('.jpg', '.jpeg', '.png'))]
```

```python
for img_file in image_files:
    try:
        print(f"Processing {img_file}...")
        results = model(img_file)  # Run YOLOv5 inference on each image

        # Print and display the results
        results.print()   # Print the results (objects detected and their confidence scores)
        results.show()    # Show the image with bounding boxes and labels
    except UnidentifiedImageError:
        print(f"Skipping corrupted or unidentified image: {img_file}")
```

```python
# Try using a sample image from the YOLOv5 repository
sample_image = '/content/yolov5/data/images/testImage.jpeg'

try:
    print(f"Processing {sample_image}...")
    results = model(sample_image)  # Run YOLOv5 inference on the sample image

    # Print and display the results
    results.print()   # Print detected objects and their confidence scores
    results.show()    # Show the image with bounding boxes and labels
except Exception as e:
    print(f"Error processing image: {e}")
```
